```python
#Ozet = csv dosyasını X ve y data setlerini ayırıyoruz. #dummies
#linearRegression ile train ediyoruz.
#model.predict([[]]) ile test ediyoruz

import pandas as pd
url = "https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/homeprices.csv"
df = pd.read_csv(url)
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies], axis = 'columns')
merged
```

```python
final = merged.drop(['town', 'west windsor'],axis='columns')
#Karisiklik olmaması icin dummies columndan 1 tanesini iptal etmek gerekiyormus
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop('price', axis = 'columns')
X
```

```python
y = final.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,y)
```

```python
##Exercise

url = 'https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/Exercise/carprices.csv'
```

```python
df = pd.read_csv(url)
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df,dummies], axis = 'columns')
merged
```

```python
final = merged.drop(["Car Model","Mercedez Benz C class"],axis='columns')
final
```

```python
X = final.drop('Sell Price($)' , axis = 'columns')
X
```

```python
y = final['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
#Price of mercedes benz that is 4yr old with mileage 45000
model.predict([[45000,4,0,0]])
```
