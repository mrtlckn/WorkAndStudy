```python
import pandas as pd
url = "https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv"

df = pd.read_csv(url)
df
```

```python
#independet variables
inputs = df.drop('salary_more_then_100k',axis ='columns')

#target variables
target = df['salary_more_then_100k']
```

```python
inputs.head()
```

```python
target.head()
```

```python
#string to numbers independet variables
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
inputs.head()
```

```python
#drop label columns
inputs_n = inputs.drop(['company', 'job', 'degree'],axis = 'columns')
inputs_n.head()
```

```python
X = inputs_n
y = target

#we not split dataset because this is simple 
#but we need to split dataset end of the day
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X,y)
```

```python
model.score(inputs_n,target)
```

```python
#Google, sales executive, masters
model.predict([[2,2,1]])
```

```python
#
#
#Exercise
#
#predict the survival rate - target Survived 
#X = pclass, gender , age , fare 

import pandas as pd

url = 'https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/Exercise/titanic.csv'

df = pd.read_csv(url)
df
```

```python
inputs = df[['Pclass','Sex', 'Age','Fare']]
#But sex class is string, we need to LabelEncoder
from sklearn.preprocessing import LabelEncoder
le_sex = LabelEncoder()
inputs['sex_n'] = le_sex.fit_transform(inputs['Sex'])
inputs

#or
#inputs.Sex = inputs.Sex.map({'male':1 , 'female':2})
```

```python
#drop String class
inputs_n = inputs.drop(['Sex'],axis = 'columns')
inputs_n
```

```python
# now we have inputs_n and target , i mean we have X and y
#but Age class have NaN columns. we need to fill it with median!
import math 
median_age = math.floor(df.Age.median())
median_age
```

```python
inputs_n.Age = df.Age.fillna(median_age)
```

```python
inputs_n
```

```python
target = df['Survived']
```

```python
#Lets build Tree and then I want to split our dataset!
from sklearn import tree
model = tree.DecisionTreeClassifier()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs_n,target,test_size = 0.2)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict([[2,28.0,7.5,1]])
```
